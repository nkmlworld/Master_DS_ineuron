**Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.**

**DataFrame:**

A DataFrame is a two-dimensional data structure in pandas.
It consists of rows and columns, similar to a spreadsheet.
Each column in a DataFrame is a Series.
DataFrames can store data of different data types in each column.

**Series**:

A Series is a one-dimensional data structure in pandas.
It consists of a sequence of values and associated labels, called indices.
Series can only store data of a single data type.

```python
import pandas as pd

# Create a DataFrame
df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Claire"],
    "Age": [25, 30, 27],
    "Gender": ["Female", "Male", "Female"],
})

# Create a Series
s = pd.Series([25, 30, 27])

# Print the DataFrame
print(df)
print("\n")
# Print the Series
print(s)
```

**Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can
you give an example of when you might use one of these functions?**

head() and tail(): These functions are used to view the first or last few rows of the DataFrame.

info() and describe(): info() gives a concise summary of the DataFrame, including the data types and number of non-null values in each column. describe() provides summary statistics for numerical columns.

iloc[] and loc[]: These functions are used for selection by label or by integer location. iloc[] is primarily integer position based, whereas loc[] is label based.
drop(): This function is used to drop specified labels from rows or columns.

fillna(): This function is used to fill missing values in the DataFrame with a specified value or method.

groupby(): This function is used to split the data into groups based on some criteria, and then apply a function to each group independently.

merge() and concat(): These functions are used to combine multiple DataFrames either by merging them based on a key column or by concatenating them along a specified axis.

apply(): This function is used to apply a function along an axis of the DataFrame.

pivot_table(): This function creates a spreadsheet-style pivot table as a DataFrame.

sort_values() and sort_index(): These functions are used to sort the DataFrame by values or index labels.

```python
import pandas as pd

# Create a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Claire', 'David', 'Emma'],
    'Age': [25, 30, 27, 32, 28],
    'Gender': ['Female', 'Male', 'Female', 'Male', 'Female']
}
df = pd.DataFrame(data)

# Group by Gender and calculate the average age
average_age_by_gender = df.groupby('Gender')['Age'].mean()

print(average_age_by_gender)
```

**Q6. Which of the following is mutable in nature Series, DataFrame, Panel?**

In Pandas, Series and DataFrame are mutable in nature, while Panel is not mutable.

Series: A Series is mutable, meaning you can modify its elements after it has been created.

DataFrame: A DataFrame is also mutable. You can add or remove columns, modify existing values, or perform various other operations that change the structure or content of the DataFrame.

Panel: Panels were a data structure in Pandas for handling three-dimensional data. However, as of version 0.25.0, Panels have been deprecated, and users are encouraged to use MultiIndex DataFrames or xarray instead. Panel objects were not mutable like Series and DataFrame.

**Q7. Create a DataFrame using multiple Series. Explain with an example.**

You can create a DataFrame using multiple Series by passing a dictionary to the pd.DataFrame() constructor. The keys of the dictionary will become the column names of the DataFrame, and the values of the dictionary will become the Series that make up the DataFrame.

For example, the following code creates a DataFrame with two columns, "Name" and "Age":

```python
import pandas as pd

name_series = pd.Series(["Alice", "Bob", "Claire"])
age_series = pd.Series([25, 30, 27])

df = pd.DataFrame({"Name":name_series, "Age":age_series})
df
```
